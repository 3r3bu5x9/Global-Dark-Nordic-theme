```python
from sympy import *
from sympy.plotting import plot3d
init_printing(use_unicode = True)
```

```python
x,y = symbols('x y')
f = x**3 - x + y**3 - y
i = .5
j = 10
```

```python
x0 = -1/(3**.5)
y0 = 1/(3**.5)
```

```python
fx = diff(f,x).subs(x,x0).subs(y,y0)
fy = diff(f,y).subs(x,x0).subs(y,y0)
fxx = diff(f,x,2).subs(x,x0).subs(y,y0)
fxy = diff(f,x,y).subs(x,x0).subs(y,y0)
fyy = diff(f,y,2).subs(x,x0).subs(y,y0)
H = Matrix([[fxx,fxy],[fxy,fyy]])
grad_f = Matrix([[fx,fy]])
dX = Matrix([x-x0,y-y0])
Q = 0.5*dX.T*H*dX
L = grad_f*dX
F = f.subs(x,x0).subs(y,y0) + Q.det() + L.det()
expand(F)
```

```python
c1=plot3d(f,(x,x0-j,x0+j),(y,y0-j,y0+j))
f
```

```python
c2=plot3d(F,(x,x0-j,x0+j),(y,y0-j,y0+j))
expand(F)
```

```python
c1.extend(c2)
c1.show()
print("together")
```

```python
plot3d(F,(x,x0-i,x0+i),(y,y0-i,y0+i))
print(f"approx at x,y = {x0},{y0}")
```

```python
plot3d(f,(x,x0-i,x0+i),(y,y0-i,y0+i))
print("original")
```
